```python
#First Forum Task
def average(numbers):
    total=float(sum(numbers))
    return (total/len(numbers))

def get_average(student):
    homework_average=average(student["homework"])
    quizzes_average=average(student["quizzes"])
    tests_average=average(student["tests"])

    homework_average=homework_average*0.1
    quizzes_average=quizzes_average*0.3
    tests_average=tests_average*0.6
    return(homework_average+quizzes_average+tests_average)

def get_letter_score(score):
    if score>=90:
        return 'A'
    elif score>=80:
        return 'B'
    elif score>=70:
        return 'C'
    elif score>=60:
        return 'D'
    else:
        return 'F'
    
def get_class_average(students):
    results=[]
    for student in students:
        student_average=get_average(student)
        results.append(student_average)
    return average(results)

#The students' dictionary
lloyd = {
 "name": "Lloyd",
 "homework": [90.0,97.0,75.0,92.0],
 "quizzes": [88.0,40.0,94.0],
 "tests": [75.0,90.0]
}
alice = {
 "name": "Alice",
 "homework": [100.0, 92.0, 98.0, 100.0],
 "quizzes": [82.0, 83.0, 91.0],
 "tests": [89.0, 97.0]
}
tyler = {
 "name": "Tyler",
 "homework": [0.0, 87.0, 75.0, 22.0],
 "quizzes": [0.0, 75.0, 78.0],
 "tests": [100.0, 100.0]
}

student_list=[lloyd,alice,tyler]

#display the dictionaries
for i in range (0,len(student_list)):
    for key,value in student_list[i].items():
        print (f"{key}: {value}")
    print('')

#call the get_average function and display the student's letter score
lloyd_average=get_average(lloyd)
print(f"Lloyd's letter score is {get_letter_score(lloyd_average)}")
alice_average=get_average(alice)
print(f"Alice's letter score is {get_letter_score(alice_average)}")
tyler_average=get_average(tyler)
print(f"Tyler's letter score is {get_letter_score(tyler_average)}")

#call the get_average function and display the class score
class_average=get_class_average(student_list)
print(f'The class average: {class_average:.3f}')
print(f'The letter score for the class: {get_letter_score(class_average)}')
```

```python
#Third Forum Task

#function to calculate the total sales
def total(number):
    total_result=0
    for i in number:
        total_result+=i
    return total_result

week=("Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday")
sales=[]

#user's input and validating the input
for day in week:
    temp=eval(input(f"Enter the sales for {day}: "))
    while(temp<0):
        temp=eval(input(f"Input was invalid. Please re-enter the sales for {day}: "))
    sales.append(temp)

#calculate the total, minimum, and maximum sales based on the user's input
total_sales=total(sales)
min_sales=min(sales)
max_sales=max(sales)

#display the output
print(f"The total sales is ${total_sales:,.2f}")
print(f"The minimum sale amount was ${min_sales:,.2f}")
print(f"The maximum sale amount was ${max_sales:,.2f}")
```
